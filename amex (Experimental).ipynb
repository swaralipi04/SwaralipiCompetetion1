```python
def giveLastNgamesPlayer(player_id, date, n, bat_or_bowl):
    '''
    Function to get last n games stats of a player before an input date.
    
    Input-
    1. player_id: id of the player to get historical data.
    2. date: date to look-back and get n games. Stats returned are before this input date.
    3. n: Number of historical games stats to return.
    4. bat_or_bowl: Kind of stats to return. {'bat': batting stats to return, 'bowl': bowling stats to return}
    
    Output-None
    
    Returns- dataframe having bowling/batting stats from last n games of a player before an input date. 
    The results are sorted by date.
    '''
    if bat_or_bowl == 'bat':
        df_topick = batsman_lvl_data
        id_col = 'batsman_id'
    else:
        df_topick = bowler_lvl_data
        id_col = 'bowler_id'
        
    return df_topick[(df_topick['match_dt']<date)&(df_topick[id_col]==float(player_id))]\
                .sort_values(by='match_dt', ascending=False).head(n)
```

```python
import pandas as pd
from tqdm import tqdm

# Assuming train_data, test_data, and batsman_lvl_data are already defined

# Filter out rows where bowler_id is NaN (indicating the batsman was not out)
dismissal_records = batsman_lvl_data[['match id', 'batsman_id', 'bowler_id', 'match_dt']].dropna(subset=['bowler_id'])

# Convert bowler_id to string to match the roster ids format
dismissal_records['bowler_id'] = dismissal_records['bowler_id'].astype(str)

# Function to count dismissals
def count_dismissals(roster_ids, opponent_roster_ids, current_date):
    dismissals = 0
    for batsman_id in roster_ids.split(':'):
        for bowler_id in opponent_roster_ids.split(':'):
            dismissals += dismissal_records[(dismissal_records['batsman_id'] == batsman_id) & 
                                             (dismissal_records['bowler_id'] == bowler_id) &
                                             (dismissal_records['match_dt'] < current_date)].shape[0]
    return dismissals


# Apply the function to train_data with progress tracking
tqdm.pandas()
train_data['team1_dismissals_against_team2'] = train_data.progress_apply(lambda x: count_dismissals(x['team1_roster_ids'], x['team2_roster_ids'], x['match_dt']), axis=1)
train_data['team2_dismissals_against_team1'] = train_data.progress_apply(lambda x: count_dismissals(x['team2_roster_ids'], x['team1_roster_ids'], x['match_dt']), axis=1)

# Apply the function to test_data with progress tracking
test_data['team1_dismissals_against_team2'] = test_data.progress_apply(lambda x: count_dismissals(x['team1_roster_ids'], x['team2_roster_ids'], x['match_dt']), axis=1)
test_data['team2_dismissals_against_team1'] = test_data.progress_apply(lambda x: count_dismissals(x['team2_roster_ids'], x['team1_roster_ids'], x['match_dt']), axis=1)

# Calculate the advantage
train_data['bowler_advantage'] = train_data['team1_dismissals_against_team2'] - train_data['team2_dismissals_against_team1']
test_data['bowler_advantage'] = test_data['team1_dismissals_against_team2'] - test_data['team2_dismissals_against_team1']

# Display the updated DataFrame
print(train_data.head())
print(test_data.head())
```

```python
def winpLastn(team_id, date, n):
    '''
    Get a team's win % in last n games. If a team has won 3 game out of their last 5, win% is 60%.
    
    Input-
    1. team_id: ID of the team to get their last n games and winner information from those games.
    2. date: match date from which to get last n historical games.
    3. n: look-back window of games.
    
    Output- None
    
    Returns- Float value denoting win% of the team in last n games.
    '''
    # filter out games with either team1/2_id as input team id, match_dt being before current game's date, sort desc by date, and get top n rows (games)
    df_rel = match_lvl_data[(match_lvl_data['match_dt']<date)&\
                      ((match_lvl_data['team1_id']==team_id)|(match_lvl_data['team2_id']==team_id))]\
                        .sort_values(by='match_dt', ascending=False).head(n) 
    win_count = df_rel[df_rel['winner_id']==team_id].shape[0] # count number of rows having winner as the input team
    if win_count == 0:
        return 0
    return round(win_count*100/df_rel.shape[0], 2) # return win% rounded to two decimal points


# In[21]:


# Compute team1's win% in last 5 games
train_data['team1_winp_last50'] = train_data.progress_apply(lambda x: \
                                  winpLastn(x['team1_id'], x['match_dt'], 50), axis=1)
# Compute team2's win% in last 5 games
train_data['team2_winp_last50'] = train_data.progress_apply(lambda x: \
                                  winpLastn(x['team2_id'], x['match_dt'], 50), axis=1)


# In[22]:


# Take the ratio of (team1's win% in their last 5 games)/(team2's win% in their last 5 games). Adding 1 to avoid divide by zero error
train_data['team_winp_last50'] = (train_data['team1_winp_last50']+1)/(train_data['team2_winp_last50']+1)
train_data.drop(columns=['team1_winp_last50', 'team2_winp_last50'], inplace=True) # drop intermediate columns


# In[23]:


train_data.shape
train_data.head(2)


# In[24]:


## Similar process for test data

test_data['team1_winp_last50'] = test_data.progress_apply(lambda x: \
            winpLastn(team_id=x['team1_id'], date=x['match_dt'], n=50), axis=1)
test_data['team2_winp_last50'] = test_data.progress_apply(lambda x: \
            winpLastn(team_id=x['team2_id'], date=x['match_dt'], n=50), axis=1)
test_data['team_winp_last50'] = (test_data['team1_winp_last50']+1)/(test_data['team2_winp_last50']+1)
test_data.drop(columns=['team1_winp_last50','team2_winp_last50'], inplace=True)
test_data.shape
test_data.head(2)
```

```python
def nostrikerateLastn(player_list, date, n):
    '''
    Function to get the total strike rate of players in the roster of a team in the last n games.
    
    Input:
    1. player_list: ':' separated list of player ids in the roster of a team.
    2. date: match date of the game to calculate this feature.
    3. n: Number of games to look back and create this feature.
    
    Output: None
    
    Returns: Sum of the strike rates of all players in the roster.
    '''
    player_list = str(player_list).split(':')  # split string of ':' separated ids into a list of ids
    res_list = []
    for player in player_list:  # loop over each player_id in the roster
        df_rel = giveLastNgamesPlayer(player_id=player, date=date, n=n, bat_or_bowl='bat')  # getting batting stats from last n games for each player.
        mean_strike_rate = df_rel['strike_rate'].mean()  # Calculate the mean strike rate for the player
        if pd.isna(mean_strike_rate):  # Check if the mean strike rate is NaN
            mean_strike_rate = 0  # Treat NaN as 0
        res_list.append(mean_strike_rate)  # Append the mean strike rate to the list
    return sum(res_list)  # Sum up values of the list which is the sum of the strike rates of all players in the roster.

# Computing the average strike rate in the last 15 games for team1 for the train dataset.
train_data['team1_strike_rate_last15'] = train_data.apply(lambda x: \
            nostrikerateLastn(player_list=x['team1_roster_ids'], date=x['match_dt'], n=15), axis=1)
# Computing the average strike rate in the last 15 games for team2 for the train dataset.
train_data['team2_strike_rate_last15'] = train_data.apply(lambda x: \
            nostrikerateLastn(player_list=x['team2_roster_ids'], date=x['match_dt'], n=15), axis=1)

# Taking the ratio of (average strike rate in the last 15 games for team1) to (average strike rate in the last 15 games for team2). Adding 1 to handle divide by zero exceptions.
train_data['team_strike_rate_last15_ratio'] = (train_data['team1_strike_rate_last15'] + 1) / (train_data['team2_strike_rate_last15'] + 1)
train_data.drop(columns=['team1_strike_rate_last15','team2_strike_rate_last15'], inplace=True) # dropping intermediate columns

# Computing the average strike rate in the last 15 games for team1 for the test dataset.
test_data['team1_strike_rate_last15'] = test_data.apply(lambda x: \
            nostrikerateLastn(player_list=x['team1_roster_ids'], date=x['match_dt'], n=15), axis=1)
# Computing the average strike rate in the last 15 games for team2 for the test dataset.
test_data['team2_strike_rate_last15'] = test_data.apply(lambda x: \
            nostrikerateLastn(player_list=x['team2_roster_ids'], date=x['match_dt'], n=15), axis=1)

# Taking the ratio of (average strike rate in the last 15 games for team1) to (average strike rate in the last 15 games for team2). Adding 1 to handle divide by zero exceptions.
test_data['team_strike_rate_last15_ratio'] = (test_data['team1_strike_rate_last15'] + 1) / (test_data['team2_strike_rate_last15'] + 1)
test_data.drop(columns=['team1_strike_rate_last15','team2_strike_rate_last15'], inplace=True) # dropping intermediate columns
```

```python
import pandas as pd

def winpLastnwithumpire(team_id, date, n, umpire1, umpire2):
    '''
    Calculate a team's win percentage in the last n games with given umpires.

    Input:
    1. team_id: ID of the team to calculate the feature for.
    2. date: Match date of the current game to calculate the feature for.
    3. n: Look-back window of games for the team.
    4. umpire1: ID of the first umpire.
    5. umpire2: ID of the second umpire.

    Output: None

    Returns: Win percentage of the team in the last n games under specified conditions.
    '''
    # Filter matches played by the team before the given date, with specified umpires
    df_rel = match_lvl_data[(match_lvl_data['match_dt'] < date) & 
                            ((match_lvl_data['team1_id'] == team_id) | (match_lvl_data['team2_id'] == team_id)) &
                            (match_lvl_data['umpire1'] == umpire1) &
                            (match_lvl_data['umpire2'] == umpire2)]\
                            .sort_values(by='match_dt', ascending=False).head(n)
    
    # Calculate the number of matches played
    total_matches = len(df_rel)
    
    # Calculate the number of matches won by the team
    matches_won = df_rel[df_rel['winner_id'] == team_id].shape[0]
    
    # Calculate the win percentage
    if total_matches == 0:
        return 0
    else:
        win_percentage = (matches_won / total_matches) * 100
        return win_percentage

def apply_winp_with_umpires(row, n, team_col):
    '''
    Extract the win percentage with umpires for each row in train_data and test_data.
    
    Input:
    1. row: A row of the DataFrame.
    2. n: Look-back window of games for the team.
    3. team_col: Column name of the team ('team1_id' or 'team2_id').
    
    Returns: Win percentage of the team in the last n games under specified conditions.
    '''
    team_id = row[team_col]
    match_info = match_lvl_data[(match_lvl_data['team1_id'] == row['team1_id']) & 
                                (match_lvl_data['team2_id'] == row['team2_id']) & 
                                (match_lvl_data['match_dt'] == row['match_dt'])]

    if not match_info.empty:
        umpire1 = match_info['umpire1'].values[0]
        umpire2 = match_info['umpire2'].values[0]
        return winpLastnwithumpire(team_id, row['match_dt'], n, umpire1, umpire2)
    else:
        return 0  # Return 0 if match information is not found

# Applying the function to train_data for team1 and team2
train_data['team1_winp_with_umpire_last15'] = train_data.apply(lambda x: apply_winp_with_umpires(x, 15, 'team1_id'), axis=1)
train_data['team2_winp_with_umpire_last15'] = train_data.apply(lambda x: apply_winp_with_umpires(x, 15, 'team2_id'), axis=1)

# Calculating the win percentage ratio for the two teams
train_data['team_winp_with_umpire_ratio_last15'] = (train_data['team1_winp_with_umpire_last15'] + 1) / (train_data['team2_winp_with_umpire_last15'] + 1)

# Dropping intermediate columns
train_data.drop(columns=['team1_winp_with_umpire_last15', 'team2_winp_with_umpire_last15'], inplace=True)

# Applying the function to test_data for team1 and team2
test_data['team1_winp_with_umpire_last15'] = test_data.apply(lambda x: apply_winp_with_umpires(x, 15, 'team1_id'), axis=1)
test_data['team2_winp_with_umpire_last15'] = test_data.apply(lambda x: apply_winp_with_umpires(x, 15, 'team2_id'), axis=1)

# Calculating the win percentage ratio for the two teams
test_data['team_winp_with_umpire_ratio_last15'] = (test_data['team1_winp_with_umpire_last15'] + 1) / (test_data['team2_winp_with_umpire_last15'] + 1)

# Dropping intermediate columns
test_data.drop(columns=['team1_winp_with_umpire_last15', 'team2_winp_with_umpire_last15'], inplace=True)
```

```python
def player_of_match_advantage(player_list, date):
    '''
    Function to calculate the player of the match advantage among items.
    
    Input:
    1. player_list: ':' separated list of player ids in the roster of a team.
    2. date: match date of the game to calculate this feature.
    
    Output: None
    
    Returns: The advantage based on the player of the match in previous games for both teams.
    '''
    player_list = [float(player) for player in str(player_list).split(':')]  # Split string of ':' separated ids into a list of ids
    
    # Get the previous games before the present date
    previous_games = match_lvl_data[match_lvl_data['match_dt'] < date]
    
    # Get the player of the match for these previous games
    previous_pom = previous_games['player_of_the_match_id']
    
    # Count the occurrences of each player being the player of the match for both teams
    team1_pom_count = sum(previous_pom.isin(player_list))
    team2_pom_count = len(previous_pom) - team1_pom_count
    
    # Calculate the advantage based on the difference in player of the match counts
    advantage = team1_pom_count - team2_pom_count
    
    return advantage
# Compute the player of the match advantage for team1 in the train dataset
train_data['team1_pom_advantage'] = train_data.apply(lambda x: \
            player_of_match_advantage(player_list=x['team1_roster_ids'], date=x['match_dt']), axis=1)

# Compute the player of the match advantage for team2 in the train dataset
train_data['team2_pom_advantage'] = train_data.apply(lambda x: \
            player_of_match_advantage(player_list=x['team2_roster_ids'], date=x['match_dt']), axis=1)

# Compute the single column representing the team's player of the match advantage in the train dataset
train_data['team_pom_advantage'] = (train_data['team1_pom_advantage'] + 1) / (train_data['team2_pom_advantage'] + 1)

# Drop the previous columns representing the advantages for each team in the train dataset
train_data.drop(columns=['team1_pom_advantage', 'team2_pom_advantage'], inplace=True)

# Compute the player of the match advantage for team1 in the test dataset
test_data['team1_pom_advantage'] = test_data.apply(lambda x: \
            player_of_match_advantage(player_list=x['team1_roster_ids'], date=x['match_dt']), axis=1)

# Compute the player of the match advantage for team2 in the test dataset
test_data['team2_pom_advantage'] = test_data.apply(lambda x: \
            player_of_match_advantage(player_list=x['team2_roster_ids'], date=x['match_dt']), axis=1)

# Compute the single column representing the team's player of the match advantage in the test dataset
test_data['team_pom_advantage'] = (test_data['team1_pom_advantage'] + 1) / (test_data['team2_pom_advantage'] + 1)

# Drop the previous columns representing the advantages for each team in the test dataset
test_data.drop(columns=['team1_pom_advantage', 'team2_pom_advantage'], inplace=True)
```

```python
def win_when_bat_or_field_lastn(team_id, date, n, ground_id, decision):
    '''
    Calculate the number of wins when a specific toss decision is made (batting or fielding)
    for the previous n matches for the specified team and ground.

    Input:
    1. team_id: ID of the team to calculate the feature for.
    2. date: Match date of the current game to calculate the feature for.
    3. n: Look-back window of games for the team and ground.
    4. ground_id: ID of the ground to consider the matches.
    5. decision: The toss decision to consider ('bat' or 'field').

    Output: None

    Returns: Number of wins when the specified toss decision is made.
    '''
    # Filter matches played by the team before the given date and at the specified ground
    df_rel = match_lvl_data[(match_lvl_data['match_dt'] < date) & 
                            ((match_lvl_data['team1_id'] == team_id) | (match_lvl_data['team2_id'] == team_id)) &
                            (match_lvl_data['ground_id'] == ground_id)].sort_values(by='match_dt', ascending=False).head(n)
    
    # Filter matches where the specified toss decision was made
    if decision == 'bat':
        df_rel = df_rel[df_rel['toss decision'] == 'bat']
    elif decision == 'field':
        df_rel = df_rel[df_rel['toss decision'] == 'field']
    
    # Calculate the number of wins when the specified toss decision was made
    if len(df_rel) == 0:
        return 0
    else:
        df_rel['won_match'] = ((df_rel['winner_id'] == team_id))
        return df_rel['won_match'].sum()

# Applying the win_when_bat_or_field_lastn function to train_data for batting decision
train_data['team_bat_field_win_ratio_last15'] = train_data.apply(lambda x: \
                (win_when_bat_or_field_lastn(x['team1_id'], x['match_dt'], 15, x['ground_id'], 'bat') + 1) / \
                (win_when_bat_or_field_lastn(x['team1_id'], x['match_dt'], 15, x['ground_id'], 'field') + 1), axis=1)

# Applying the win_when_bat_or_field_lastn function to test_data for batting decision
test_data['team_bat_field_win_ratio_last15'] = test_data.apply(lambda x: \
                (win_when_bat_or_field_lastn(x['team1_id'], x['match_dt'], 15, x['ground_id'], 'bat') + 1) / \
                (win_when_bat_or_field_lastn(x['team1_id'], x['match_dt'], 15, x['ground_id'], 'field') + 1), axis=1)
```

```python
# Function to calculate average wins by wicket or runs for a ground in the last n games
def avg_by_wicket_or_run_Ground(ground_id, date, n):
    '''
    Function to calculate average wins by wicket or runs
    
    Input:
    1. ground_id: ID of the ground to calculate the feature for.
    2. date: Match date of the current game to calculate the feature for.
    3. n: Look-back window of games for the ground.

    Output: None (Modify as needed)

    Returns: Difference between average wins by wickets and average wins by runs for the specified ground and look-back window.
    '''
    # Filter games played at the input ground_id before the input date. Sort by date and select top n rows.
    df_rel = match_lvl_data[(match_lvl_data['match_dt'] < date) & (match_lvl_data['ground_id'] == ground_id)] \
             .sort_values(by='match_dt', ascending=False).head(n)
    
    # Calculate the number of wins by wickets and by runs for the specified ground
    wins_by_wickets = df_rel[df_rel['by'] == 'wickets'].shape[0]
    wins_by_runs = df_rel[df_rel['by'] == 'runs'].shape[0]
    
    # Calculate the difference between the two
    difference = wins_by_wickets - wins_by_runs
    
    return difference

# Apply the function to train_data
train_data['ground_avg_wins_by_wicket_or_run_last15'] = train_data.apply(lambda x: avg_by_wicket_or_run_Ground(x['ground_id'], x['match_dt'], 15), axis=1)

# Apply the function to test_data
test_data['ground_avg_wins_by_wicket_or_run_last15'] = test_data.apply(lambda x: avg_by_wicket_or_run_Ground(x['ground_id'], x['match_dt'], 15), axis=1)
```

```python
# Function to determine if it's a bowling-friendly ground or batting-friendly ground
def ground_type(ground_avg_wins):
    if ground_avg_wins > 0:
        return 1
    else:
        return 0

# Apply the function to create a new column indicating the ground type
train_data['ground_type'] = train_data['ground_avg_wins_by_wicket_or_run_last15'].apply(ground_type)
test_data['ground_type'] = test_data['ground_avg_wins_by_wicket_or_run_last15'].apply(ground_type)

# Determine the advantageous decision based on toss and quantify the advantage
def toss_advantage(toss_decision, ground_type):
    if toss_decision == 'field' and ground_type == 1:
        return 1  # Advantage for bowling first
    elif toss_decision == 'bat' and ground_type == 0:
        return -1  # Advantage for batting first
    else:
        return 0  # No clear advantage

# Apply the function to determine the advantageous decision and quantify the advantage
train_data['advantage'] = train_data.apply(lambda x: toss_advantage(x['toss decision'], x['ground_type']), axis=1)
test_data['advantage'] = test_data.apply(lambda x: toss_advantage(x['toss decision'], x['ground_type']), axis=1)
```

```python
def count_batsman_bowler_ratio(player_list, date, n):
    '''
    Function to calculate the batsman-to-bowler ratio for a given roster of players in the last n games.
    
    Input:
    1. player_list: ':' separated list of player ids in the roster of a team.
    2. date: match date of the game to calculate this feature.
    3. n: Number of games to look back and create this feature.
    
    Output:
    The batsman-to-bowler ratio for the given roster.
    '''
    player_list = str(player_list).split(':')  # Split string of ':' separated ids into a list of ids
    batsman_count = 0
    bowler_count = 0
    for player in player_list:  # Loop over each player_id in the roster
        # Assuming you have a function giveLastNgamesPlayer() to get player's data for the last n games
        df_rel = giveLastNgamesPlayer(player_id=player, date=date, n=n, bat_or_bowl='bat')  
        if df_rel is not None:
            # If the player is a batsman, increment the batsman_count
            batsman_count += 1
        else:
            # If the player is a bowler, increment the bowler_count
            bowler_count += 1
    # Calculate the ratio
    ratio = batsman_count / bowler_count if bowler_count != 0 else batsman_count
    return ratio

# Calculate the ratio for team1 in the last 15 games for the train dataset
train_data['team1_batsman_bowler_ratio'] = train_data.apply(lambda x: 
            count_batsman_bowler_ratio(player_list=x['team1_roster_ids'], date=x['match_dt'], n=15), axis=1)

# Calculate the ratio for team2 in the last 15 games for the train dataset
train_data['team2_batsman_bowler_ratio'] = train_data.apply(lambda x: 
            count_batsman_bowler_ratio(player_list=x['team2_roster_ids'], date=x['match_dt'], n=15), axis=1)

# Compute the difference between team1 and team2 ratios
train_data['batsman_bowler_ratio_difference'] = (train_data['team1_batsman_bowler_ratio']+0.1)/(0.1+train_data['team2_batsman_bowler_ratio'])


# Calculate the ratio for team1 in the last 15 games for the test dataset
test_data['team1_batsman_bowler_ratio'] = test_data.apply(lambda x: 
            count_batsman_bowler_ratio(player_list=x['team1_roster_ids'], date=x['match_dt'], n=15), axis=1)

# Calculate the ratio for team2 in the last 15 games for the test dataset
test_data['team2_batsman_bowler_ratio'] = test_data.apply(lambda x: 
            count_batsman_bowler_ratio(player_list=x['team2_roster_ids'], date=x['match_dt'], n=15), axis=1)

# Compute the difference between team1 and team2 ratios for the test dataset
test_data['batsman_bowler_ratio_difference'] = (test_data['team1_batsman_bowler_ratio']+0.1)/(0.1+test_data['team2_batsman_bowler_ratio'])
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import learning_curve

# Define a function to plot learning curves
def plot_learning_curve(estimator, X, y, title):
    train_sizes, train_scores, test_scores = learning_curve(
        estimator, X, y, train_sizes=np.linspace(0.1, 1.0, 10), cv=10, scoring='accuracy', n_jobs=-1)
    
    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)

    plt.figure()
    plt.title(title)
    plt.xlabel("Training examples")
    plt.ylabel("Score")
    plt.grid()

    # Increase the number of y-axis grids by specifying y-ticks
    y_min = min(train_scores_mean.min(), test_scores_mean.min())
    y_max = max(train_scores_mean.max(), test_scores_mean.max())
    y_ticks = np.linspace(y_min, y_max, 20)  # Adjust the number here for more or fewer grids
    plt.yticks(y_ticks)

    plt.fill_between(train_sizes, train_scores_mean - train_scores_std,
                     train_scores_mean + train_scores_std, alpha=0.1,
                     color="r")
    plt.fill_between(train_sizes, test_scores_mean - test_scores_std,
                     test_scores_mean + test_scores_std, alpha=0.1,
                     color="g")
    plt.plot(train_sizes, train_scores_mean, 'o-', color="r", label="Training score")
    plt.plot(train_sizes, test_scores_mean, 'o-', color="g", label="Cross-validation score")
    plt.legend(loc="best")
    plt.show()

# Plot learning curves
plot_learning_curve(clf_gbm, X, y, title="Learning Curves (CatBoostClassifier)")
```
